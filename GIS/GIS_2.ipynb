```python
import os
import rasterio
from rasterio.crs import CRS
import numpy as np
import matplotlib.pyplot as plt
import geemap
import ee
import geopandas
```

```python
overijssel_places = geopandas.read_file("./data/overijssel-latest-free.shp/gis_osm_places_free_1.shp")
overijssel_railways = geopandas.read_file("./data/overijssel-latest-free.shp/gis_osm_railways_free_1.shp")   
```

```python
print("Places Dataset:")
print(overijssel_places.crs)  
print(overijssel_places.head()) 

print("\nRailways Dataset:")
print(overijssel_railways.crs)
print(overijssel_railways.head())
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
overijssel_places.plot(ax=ax, markersize=5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
overijssel_railways.plot(ax=ax)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
overijssel_railways.plot(ax=ax)
overijssel_places.plot(ax=ax, markersize=5)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.grid(True)
plt.show()
```
